```python
import numpy as np
import matplotlib.pyplot as plt
from kececicurve import KececiCurve

# Ortak parametreler (istediğiniz gibi değiştirebilirsiniz)
params = {
    'num_children': 4,
    'max_level': 3,
    'scale_factor': 0.45,
    'base_radius': 1.8,
    'angle_variation': 0.2,
    'growth_mode': 'outward',
}

# Üç farklı stil
curves = [
    KececiCurve(**params, inherit_parent_angle=False),   # klasik (bağımsız)
    KececiCurve(**params, inherit_parent_angle=True),    # spiral (bağımlı)
    KececiCurve(num_children=5, max_level=2, growth_mode='outward')  # farklı parametrelerle
]

# Yan yana 3 subplot
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
titles = ["Klasik (bağımsız açı)", "Spiral (bağımlı açı)", "Farklı parametreler"]

for ax, curve, title in zip(axes, curves, titles):
    points = np.array(curve.generate())
    ax.plot(points[:,0], points[:,1], 'b-', lw=0.8)
    ax.scatter(points[:,0], points[:,1], c='r', s=10, zorder=5)
    # Otomatik başlık: parametreleri göster
    ax.set_title(f"{title}\n"
                 f"n={curve.num_children}, l={curve.max_level}, "
                 f"scale={curve.scale_factor}, angVar={curve.angle_variation}\n"
                 f"inherit={curve.inherit_parent_angle}", fontsize=8)
    ax.axis('equal')
    ax.axis('off')

plt.tight_layout()
plt.show()
```
